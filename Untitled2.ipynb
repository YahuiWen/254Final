```python
def get_batches(image,label,resize_w,resize_h,batch_size,capacity):
    #convert the list of images and labels to tensor
    image = tf.cast(image,tf.string)
    label = tf.cast(label,tf.float32)
    queue = tf.train.slice_input_producer([image,label])
    label = queue[1]
    image_c = tf.read_file(queue[0])
    #Decode a PNG-encoded image , channels: output an RGB image.
    image = tf.image.decode_png(image_c,channels = 3)
    #resize
    #Crops and/or pads an image to a target width and height.
    image = tf.image.resize_image_with_crop_or_pad(image,resize_w,resize_h)
    #(x - mean) / adjusted_stddev  and return the standardized image with same shape as image
    image = tf.image.per_image_standardization(image)
    #batch_size: The new batch size pulled from the queue.
    #capacity: An integer. The maximum number of elements in the queue
    image_batch,label_batch = tf.train.batch([image,label],
            batch_size = batch_size,
            num_threads = 64,
            capacity = capacity)
#     images_batch = tf.cast(image_batch,tf.string)
#     labels_batch = tf.reshape(label_batch,[batch_size])
    return image_batch,label_batch
```

```python
def model(images,batch_size):
    #defining ops that creates variables (layers).
    with tf.variable_scope('conv1') as scope:
        #tf.truncated_normal_initializer: the recommended initializer for neural network weights and filters.
        weights = tf.get_variable('weights',shape = [3,3,3, 16],dtype = tf.float32,initializer=tf.truncated_normal_initializer(stddev=0.1,dtype=tf.float32))
        biases = tf.get_variable('biases',shape=[16],dtype=tf.float32,initializer=tf.constant_initializer(0.1))
        #Computes a 2-D convolution
        conv = tf.nn.conv2d(images, weights, strides=[1,1,1,1], padding='SAME')
        #add bias to value
        pre_activation = tf.nn.bias_add(conv, biases)
        conv1 = tf.nn.relu(pre_activation, name= scope.name)
    with tf.variable_scope('pooling1_lrn') as scope:
        #Performs the max pooling on the input.
        pool1 = tf.nn.max_pool(conv1, ksize=[1,2,2,1],strides=[1,2,2,1],padding='SAME', name='pooling1')
        #Local Response Normalization.
        norm1 = tf.nn.lrn(pool1, depth_radius=4, bias=1.0, alpha=0.001/9.0,beta=0.75,name='norm1')
    with tf.variable_scope('conv2') as scope:
        weights = tf.get_variable('weights',shape=[3,3,16,128],dtype=tf.float32,initializer=tf.truncated_normal_initializer(stddev=0.1,dtype=tf.float32))
        biases = tf.get_variable('biases',shape=[128],dtype=tf.float32,initializer=tf.constant_initializer(0.1))
        conv = tf.nn.conv2d(norm1, weights, strides=[1,1,1,1],padding='SAME')
        pre_activation = tf.nn.bias_add(conv, biases)
        conv2 = tf.nn.relu(pre_activation, name='conv2')
    with tf.variable_scope('pooling2_lrn') as scope:
        norm2 = tf.nn.lrn(conv2, depth_radius=4, bias=1.0, alpha=0.001/9.0,beta=0.75,name='norm2')
        pool2 = tf.nn.max_pool(norm2, ksize=[1,2,2,1], strides=[1,1,1,1],padding='SAME',name='pooling2')
    with tf.variable_scope('local3') as scope:
        reshape = tf.reshape(pool2, shape=[batch_size, -1])
        dim = reshape.get_shape()[1].value
        weights = tf.get_variable('weights',shape=[dim,1024],dtype=tf.float32,initializer=tf.truncated_normal_initializer(stddev=0.005,dtype=tf.float32))
        biases = tf.get_variable('biases',shape=[1024],dtype=tf.float32,initializer=tf.constant_initializer(0.1))
        local3 = tf.nn.tanh(tf.matmul(reshape, weights) + biases, name=scope.name)
    with tf.variable_scope('softmax_linear') as scope:
        weights = tf.get_variable('softmax_linear',shape=[1024, 1],dtype=tf.float32,initializer=tf.truncated_normal_initializer(stddev=0.005,dtype=tf.float32))
        biases = tf.get_variable('biases',shape=[1],dtype=tf.float32,initializer=tf.constant_initializer(0.1))
        softmax_linear = tf.add(tf.matmul(local3, weights),biases, name='softmax_linear')
        tf.add_to_collection('network-output', softmax_linear)
    return softmax_linear
```

```python
def loss(logits, label_batches):
    label_batches=tf.reshape(label_batches,[2,1])
#     print(label_batches.shape)
#     cost = tf.losses.mean_squared_error(logits, label_batches)
    cost = tf.losses.mean_squared_error(logits, label_batches)
#     cost = tf.reduce_mean(-tf.reduce_sum(label_batches*tf.log(logits), reduction_indices=1))
    tf.add_to_collection('cost',cost)
    return cost
```

```python
def run_training(image_batches,label_batches):
    loss_list=[]
    fitted_list=[]
    log_dir = './'
    p = model(image_batches,2)
    fitted_list.append(p)
    cost = loss(p,label_batches)
    train_op = training(cost,0.005)
    # acc = get_accuracy(p,label_batches)

    sess = tf.Session()
    init = tf.global_variables_initializer()
    sess.run(init)
    saver = tf.train.Saver()
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess = sess,coord = coord)
    epoches=11
    try:
        for epoch in range(epoches):
            for step in np.arange(2559):

                if coord.should_stop():
                    break
                _,train_loss = sess.run([train_op,cost])
                loss_list.append(train_loss)
            print("loss:{} ".format(train_loss))
            if epoch % 10 == 0:
                check = os.path.join(log_dir,"model.ckpt")
                saver.save(sess,check,global_step = epoch)
    except tf.errors.OutOfRangeError:
        print("Done!!!")
    finally:
        coord.request_stop()
    coord.join(threads)
    sess.close()
    return loss_list,fitted_list
```
